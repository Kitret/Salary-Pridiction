```python
# function to implement different models use later
def apply_model(model,x_train,y_train,x_test,y_test):
    m=model
    m.fit(x_train,y_train)
    print("TRAINING SCORE: ", m.score(x_train,y_train),'\n')
    print('TEST SCORE: ', m.score(x_test,y_test),'\n')
    y_pred = m.predict(x_test)
    print('YPRED\n', y_pred,'\n')
    cm = confusion_matrix(y_test,y_pred)
    print('CONFUSION MATRIX: \n', cm,'\n')
    print("CLASSIFICATION REPORT: \n", classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,0:-1]
y = df.iloc[:,-1]
print(x.shape)
print(y.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
gnb = GaussianNB()
lg = LogisticRegression(solver='liblinear')
dt = DecisionTreeClassifier(criterion='entropy')
knn = KNeighborsClassifier(n_neighbors=10)
rf = RandomForestClassifier(n_estimators=70,criterion='gini',max_depth=7)
```

```python
# APPLYING LINEAR REGRESSION MODEL
apply_model(lg,x_train,y_train,x_test,y_test)
```

```python
# APPLYING DESICION TREE MODEL
apply_model(dt,x_train,y_train,x_test,y_test)
```

```python
# APPLYING KNN MODEL
apply_model(knn,x_train,y_train,x_test,y_test)
```

```python
# APPLYING RANDOM FOREST MODEL
apply_model(rf,x_train,y_train,x_test,y_test)
```

```python
# APPLYING GAUSSIAN NAIVE BAYES MODEL
apply_model(gnb,x_train,y_train,x_test,y_test)
```
